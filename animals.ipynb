```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Animal_Shelter_Intake_and_Outcome.csv")
df.info()
```

```python
# heatmap
pivot_table = df.pivot_table(index='Type', columns='Intake Type', aggfunc='size', fill_value=0)
plt.figure(figsize=(10, 8)) 
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt="d")
plt.title("Animal Shelter Intake and Outcome Heatmap")
plt.xlabel("Intake Type")
plt.ylabel("Type")
plt.show()
```

```python
type_counts = df['Type'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=type_counts.index, y=type_counts.values)
plt.title("Animal Shelter Intake and Outcome Bar Chart")
plt.xlabel("Type")
plt.ylabel("intake and outcome")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Sex'], kde=True)
plt.title("Distribution of all animals sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```
